```python
import pandas as pd

df = pd.read_csv("../data/website_wata.csv")

df.head()
df.info()
df.describe()
```

```python
df["Session Duration"].describe()
```

```python
df["Page Views"].describe()
```

```python
import matplotlib.pyplot as plt

# Session Duration
plt.figure(figsize=(8, 5))
plt.hist(df["Session Duration"], bins=30)
plt.xlabel("Session Duration (min)")
plt.ylabel("Number of sessions")
plt.title("Distribution of Session Duration (minutes)")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
plt.savefig("../outputs/session_duration_hist.png", dpi=120)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Page Views"], bins=range(0, df["Page Views"].max() + 2))
plt.title("Distribution of Page Views per Session")
plt.xlabel("Page Views")
plt.ylabel("Number of sessions")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()

plt.savefig("../outputs/page_views_hist.png", dpi=120)

plt.show()
plt.close()
```
